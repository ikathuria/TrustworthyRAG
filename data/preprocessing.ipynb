# DocBench

```python
import os
from glob import glob
import pandas as pd
from tqdm import tqdm
```

```python
doc_bench_folders = [f for f in glob('raw/DocBench/*') if os.path.isdir(f)]
len(doc_bench_folders)
```

```python
result_df = pd.DataFrame()
for folder in tqdm(doc_bench_folders):
	try:
		# we have  pdf file and a jsonl file with question/answer pairs for the pdf
		pdf_file = glob(os.path.join(folder, "*.pdf"))[0]
		jsonl_file = glob(os.path.join(folder, "*.jsonl"))[0]

		# load the jsonl file into a dataframe
		df = pd.read_json(jsonl_file, lines=True)

		# add a column for the pdf file path
		df["pdf_file"] = pdf_file

		result_df = pd.concat([result_df, df], ignore_index=True)
	except Exception as e:
		print(f"Error processing folder {folder}: {e}")
		raise e
```
